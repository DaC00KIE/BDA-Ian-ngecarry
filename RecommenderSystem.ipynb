# Setup

```python
# Import Library
import pandas as pd
```

## Functions

```python
# Split nutrition function
def split_nutrition(nutrition):
    if not isinstance(nutrition, str):
        return {}
    
    nutritions = {}
    try:
        items = nutrition.split(", ")
        for item in items:
            # Attempt to split into key-value pairs
            if ": " in item:
                key, value = item.split(": ", 1)
                nutritions[key.strip()] = value.strip()
    except Exception as e:
        print(f"Error parsing nutrition: {nutrition} - {e}")
    return nutritions
```

## All Recipes

```python
# Load Dataset Allrecipes
allrecipes_df = pd.read_csv('AllRecipes_Scraper/AllRecipesDataset.csv')
allrecipes_df.head()
```

```python
# Split nutrition
nutrition_split = allrecipes_df["nutrition"].apply(split_nutrition)
nutrition_split= pd.DataFrame(list(nutrition_split))
allrecipes_df = pd.concat([allrecipes_df, nutrition_split], axis=1)

allrecipes_df.head()
```

```python
# Drop nutrition column
allrecipes_df = allrecipes_df.drop('nutrition', axis=1)
allrecipes_df.head()
```

## BBCGoodFood

```python
# Load Dataset BBCGoodFood
bbc_df = pd.read_csv('BBCGoodFood_Scraper/BBCGoodFoodScrap.csv')
bbc_df.head()
```

```python
# Split nutrition
nutrition_split = bbc_df["nutrition"].apply(split_nutrition)
nutrition_split= pd.DataFrame(list(nutrition_split))
bbc_df = pd.concat([bbc_df, nutrition_split], axis=1)

bbc_df.head()
```

```python
# Drop saturates, sugars, fibre, salt
bbc_df = bbc_df.drop(columns=["'fibre'", "'saturates'", "'salt'", "'sugars'", "nutrition"])
bbc_df.head()
```

```python
# Rename the columns
bbc_df = bbc_df.rename(columns={
    "title" : "name",
    "step" : "steps",
    "({'kcal'" : "calories",
    "'fat'" : "fat",
    "'carbs'" : "carbs",
    "'protein'" : "protein",
    "serve" : "servings"
})

bbc_df.head()
```

## Taste

```python
# Load Dataset Taste
taste_df = pd.read_csv('TasteScrape/tastescrapedata2.csv')
taste_df.head()
```

```python
# Drop sodium, fibre and unnamed: 0
taste_df = taste_df.drop(columns=['Unnamed: 0', 'sodium', 'fibre'])
taste_df.head()
```

```python
# Rename the columns
taste_df = taste_df.rename(columns={
    "foodname" : "name",
    "url" : "link",
    "image" : "image_url",
    "carb" : "carbs",
    "serving" : "servings",
    "ingredient" : "ingredients",
})

taste_df.head()
```

## Merge CSV

```python
# Re-order the attributes
new_order = ['name', 'ingredients', 'steps', 'calories', 'carbs', 'protein', 'fat', 'servings', 'image_url', 'link']

# AllRecipes
allrecipes_df = allrecipes_df[new_order]
allrecipes_df.head()
```

```python
# BBCGoodFood
bbc_df = bbc_df[new_order]
bbc_df.head()
```

```python
# Taste
taste_df = taste_df[new_order]
taste_df.head()
```

```python
# Merge by a common key column
merged_df = pd.concat([allrecipes_df, bbc_df, taste_df], ignore_index=True)

# Save to a new CSV
merged_df.to_csv("Food_Dataset.csv", index=False)
```
